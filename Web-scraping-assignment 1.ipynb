### Write a python program to display IMDB’s Top rated 100 movie's data (i.e.name,rating,year of release) and make data frame.

```python
#url = 'http://www.imdb.com/chart/top'
def get_soup(txt):
    soup = BeautifulSoup(get_url(txt).text, 'lxml')
    return soup
```

```python
def make_data_list(soup):
    movies = soup.select('td.titleColumn')
    links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
    crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
    ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]
    votes = [b.attrs.get('data-value') for b in soup.select('td.ratingColumn strong')]

    lis = []
 
    for index in range(0, len(movies)):  
    # Separating  movie into: 'place',
    # 'title', 'year'
        movie_string = movies[index].get_text()
        movie = (' '.join(movie_string.split()).replace('.', ''))
        movie_title = movie[len(str(index))+1:-7]
        year = re.search('\((.*?)\)', movie_string).group(1)
        place = movie[:len(str(index))-(len(movie))]
     
        data = {"Name": movie_title,
            "Year of Release": year,
            "Rating": ratings[index],}
        lis.append(data)
    return lis
```

```python
def make_display(lis):
    df = pd.DataFrame(lis)
    return df
```

```python
def get_table(soup):
    return soup.find('table')
```

```python
def get_rows(table):
    return table.find_all('tr')
```

```python
def return_rating_movies(row):
    lis = []
    temp = row.text.strip().split('\n')
    for n in range(0,len(temp)):
        if temp[n] != '':
            lis.append(temp[n])
    return [lis[1].strip(),lis[2],lis[3]]    
```

```python
def make_rating_table(url):
    final_list = []
    table = get_table(get_soup(url))
    rows = get_rows(table)
    for i in range(1,11):
        final_list.append(return_rating_movies(rows[i]))
    
    df = pd.DataFrame(final_list)
    df.rename(columns = {0:'Title', 1:'Year of release',2:'Rating'}, inplace = True)
    return df
```

```python
(make_rating_table('https://www.imdb.com/india/top-rated-indian-movies/'))
```

```python
def get_card(soup):
    return soup.find_all('div', class_ = 'Card__BaseCard-sc-b3n78k-0 lfjhF NewProductCard__CardStyled-sc-j0e7tu-0 hzPSVW NewProductCard__CardStyled-sc-j0e7tu-0 hzPSVW')
```

```python
def get_subcard(card):
    return card.find('div',class_ = 'Card__BaseCard-sc-b3n78k-0 cXRroa NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 eEBPAI NewProductCard__DetailCard_Desktop-sc-j0e7tu-2 eEBPAI')
```

```python
def get_product_name(subcard):
    return subcard.find('p', class_ = 'Text__StyledText-sc-oo0kvp-0 cPgaBh NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 hofZGw NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 hofZGw').text
```

```python
def get_discounted_price(subcard):
    return subcard.find('h5').text
```

```python
def get_discount_percentage(subcard):
    return subcard.find('span', color = 'greenBase').text
```

```python
def make_product_list(subcard):
    return [get_product_name(subcard), get_discounted_price(subcard), get_discount_percentage(subcard)]    
```

```python
def make_product_table(url):
    product_list = []
    soup = get_soup(url)
    card = get_card(soup)
    
    for i in range(0,len(card)):
        subcard = get_subcard(card[i])
        product_list.append(make_product_list(subcard))
    df = pd.DataFrame(product_list)
    df.rename(columns = {0:'Product Name', 1:'Discounted Price',2:'Discount Rate'}, inplace = True)
    
    return df
        
```

```python
make_product_table('https://meesho.com/bags-ladies/pl/p7vbp')
```

```python
def get_Team_ODI_List(soup):
    results = soup.find_all("td")
    lis = []
    for i in range(1,len(results),5):
        if i == len(results)-1:
            break
        else: 
            name = results[i].text.strip().split(".")[-1].strip().split("\n")[0]
            matches = results[i+1].text#strip().split(".")[-1].strip().split("\n")[-1][1:-1]
            points = results[i+2].text
            rating = results[i+3].text.strip()
        lis.append([name ,matches ,points ,rating])
    return lis
```

```python
def get_headline(article):
    return article.h2.a.text
```

```python
def get_date(article):
    return article.find('time', class_ = 'entry-time').text
```

```python
def get_summary(article):
    return article.find('div',class_ = 'entry-content').text.strip()
```

```python
def get_video_Code(article):
    yt = article.find('iframe',class_ = 'youtube-player')
    code = yt['src'].split('/')[4].split('?')[0]
    return code
```

```python
def get_list(article):
    return [get_headline(article), get_date(article), get_summary(article), get_video_Code(article)]
```

```python
def make_table(url):
    final_L = []
    soup = get_soup(url)
    articles = soup.find_all('article')
    for i in range(0,len(articles)):
        if i == 4:
            continue
        else:
            final_L.append(get_list(articles[i]))
    df = pd.DataFrame(final_L)
    df.rename(columns = {0:'Title', 1:'Date',2:'Summary',3: 'Youtube Code'}, inplace = True)
    
    return df           
```

```python
make_table('https://coreyms.com/')
```

```python
def get_name(tile):
    return tile.find('a',class_ = 'restnt-name ellipsis').text
```

```python
def get_location(tile):
    return tile.find('div', class_ = 'restnt-loc ellipsis').text
```

```python
def get_cuisines(tile):
    c = tile.find('div', class_ = 'detail-info')
    return (c.text.split("|")[1])
```

```python
def get_rating(tile):
    return tile.find('div', class_= 'restnt-rating rating-4').text
```

```python
def get_image_url(tile):
    return tile.find('img', class_= 'lazy-load-img no-img')['data-src']
```

```python
def get_list_restnt(tile):
    return [get_name(tile), get_location(tile), get_cuisines(tile), get_rating(tile), get_image_url(tile)]
```

```python
def make_restnt_table(url):
    final_L = []
    soup = get_soup(url)
    tiles = soup.find_all('div', class_ = 'restnt-main-wrap clearfix')
    for i in range(0,len(tiles)):
        final_L.append(get_list_restnt(tiles[i]))
    df = pd.DataFrame(final_L)
    df.rename(columns = {0:'Name', 1:'Location',2:'Cuisines',3: 'Rating', 4: 'Image URL'}, inplace = True)
    
    return df
```

```python
make_restnt_table('https://www.dineout.co.in/delhi-restaurants/buffet-special')
```
