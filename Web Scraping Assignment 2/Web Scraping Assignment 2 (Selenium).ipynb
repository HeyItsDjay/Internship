>2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the

>5. Finally create a dataframe of the scraped data.

<br><b>Note</b>: All of the above steps have to be done in code. No step is to be done manually.

```python
#Connecting to Naukri.com
driver.get('http://www.naukri.com/')
```

```python
#Searching for Job Position - Data Analyst
designation = driver.find_element(By.CLASS_NAME, "suggestor-input")
designation.send_keys("Data Analyst")
```

```python
#Searching for Job Location - Banglore
location = driver.find_element(By.XPATH, "/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys("Bangalore")
```

```python
#Clicking on the Submit button
search = driver.find_element(By.CLASS_NAME, "qsbSubmit")
search.click()
```

```python
job_title = []
job_location = []
company_name = []
experience_required = []
```

```python
title_tags = driver.find_elements(By.XPATH, '//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    job_title.append(i.text)

location_tags = driver.find_elements(By.XPATH, '//span[@class="ellipsis fleft locWdth"]')
for i in location_tags[0:10]:
    job_location.append(i.text)

name_tags = driver.find_elements(By.XPATH, '//a[@class="subTitle ellipsis fleft"]')
for i in name_tags[0:10]:
    company_name.append(i.text)
    
exp_tags = driver.find_elements(By.XPATH, '//span[@class="ellipsis fleft expwdth"]')
for i in exp_tags[0:10]:
    experience_required.append(i.text)
```

```python
#Compiling the data
cols = ['Job Title', 'Location', 'Company Name', 'Experience Required']
data = list(zip(job_title, job_location, company_name, experience_required))

#Creating a dataframe
df = pd.DataFrame(data= data, columns = cols)
df
```

<span style="font-family:Calibri; font-size:2em;"><b>Question 4</b></span>

Scrape data of first 100 sunglasses listings on <A href = 'Flipkart.com'>Flipkart</A>. You have to scrape four attributes:
<br>
1. Brand
2. Product Description
3. Price
4. Discount

<br>The attributes which you have to scrape is ticked marked in the below image -
<br>


<br>

```python
#Connecting to Flipkart
driver.get('http://www.flipkart.com/')
```

```python
#Searching for Sunglasses
Product = driver.find_element(By.CLASS_NAME, "_3704LK")
Product.send_keys("Sunglasses")
```

```python
#Clicking on Search Button
Search_Button = driver.find_element(By.XPATH,"/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
Search_Button.click()
```

```python
start = 0
end = 3
n = 40
Brand = []
Desc = []
Price = []
Disc = []

for page in range(start, end):
    if page == 2:
        n = 20         
    Name_Tags = driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
    for i in Name_Tags[0:n]:
        Brand.append(i.text)

    Desc_Tags = driver.find_elements(By.XPATH, '//a[@class="IRpwTa"] | //a[@class="IRpwTa _2-ICcC"]')
    for i in Desc_Tags[0:n]:
        Desc.append(i.text)

    Price_Tags = driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
    for i in Price_Tags[0:n]:
        Price.append(i.text)

    Disc_Tags = driver.find_elements(By.XPATH, '//div[@class="_3Ay6Sb"]')
    for i in Disc_Tags[0:n]:
        Disc.append(i.text.split()[0])

    Next_Button = driver.find_element(By.CLASS_NAME, "_1LKTO3")
    Next_Button.click()
        
    time.sleep(3)       
```

```python
#Consolidating the data
cols = ['Brand', 'Description', 'Price', 'Discount']
data = list(zip(Brand, Desc, Price, Disc))

#Creating the Dataframe
df = pd.DataFrame(data= data, columns = cols)
df
```

<span style="font-family:Calibri; font-size:2em;"><b>Question 8</b></span>

Write a python program to scrape data for <b>Top 1000 Quotes of All Time</b>.
<br>
<br>The above task will be done in following steps:
1. First get the webpage - <a href = 'https://www.azquotes.com/'>AZQuotes</a>
2. Click on Top Quotes
3. Then scrap - a) Quote b) Author c) Type Of Quotes
<br>



<br>
<br>

```python
#Connecting to AZQuotes.com
driver.get("http://www.azquotes.com/")
```

```python
#Navigating to Top Quotes
Top_Quotes = driver.find_element(By.XPATH, "/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a")
Top_Quotes.click()
```

```python
start = 0
end = 10

Quote = []
Author = []
Type = []

for page in range(start, end):
    Quote_Tags = driver.find_elements(By.XPATH, '//a[@class="title"]')
    for i in Quote_Tags:
        Quote.append(i.text)

    Author_Tags = driver.find_elements(By.XPATH, '//div[@class="author"]')
    for i in Author_Tags:
        Author.append(i.text)

    Type_Tags = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in Type_Tags:
        Type.append(i.text)
        
    if page==10:
        break
        
    time.sleep(3)
```

```python
#Consolidating the Data
cols = ['Quote', 'Author', 'Type']
data = list(zip(Quote, Author, Type))

#Creating the Dataframe
df = pd.DataFrame(data= data, columns = cols)
df
```

```python
#Connecting to Motor1 
driver.get("https://www.motor1.com/")
```

```python
#Searching for 50 Most Expensive Cars
Search_Bar = driver.find_element(By.XPATH, '/html/body/div[3]/div[2]/div/div/div[3]/div/div/button')
Search_Bar.send_keys('50 most expensive cars')
```

```python
#Clicking the Search Button
Search_Button = driver.find_element(By.XPATH, '/html/body/div[3]/div[2]/div/div/div[3]/div/div/div/form/button[1]')
Search_Button.click()
```

```python
#Clicking the link
Link = driver.find_element(By.XPATH, '/html/body/div[3]/div[9]/div[4]/div[1]/div/div/div[4]/div/div[1]/h3/a')
Link.click()
```

```python
Name = []
Price = []
Desc = []

Name_Tags = driver.find_elements(By.XPATH, '//h3[@class="subheader"]')
for i in Name_Tags[0:50]:
    Name.append(i.text)
    Price.append(i.find_element(By.XPATH, "following-sibling::*[2]").text.replace('Price:',''))
    Desc.append(i.find_element(By.XPATH, "following-sibling::*[3]").text)
```
